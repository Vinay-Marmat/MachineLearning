```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,50,25)
```

```python
y=x*x
y1= y*x
y2= y1*x
y3= y2*x
y4= 1/x
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y1,'r--')
plt.subplot(2,2,3)
plt.plot(x,y3,'bo')
plt.subplot(2,2,4)
plt.plot(x,y4,'b*-')
```

```python
x= np.arange(00,4*np.pi,0.1)
x2= np.linspace(0,4*np.pi,100)
```

```python
y = np.sin(x)
y2= np.sin(x2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Plot of sin(x)')
plt.subplot(1,2,2)
plt.plot(x2,y2,'g--')
```

```python
x=np.linspace(0,4*np.pi,1000)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
grid = plt.GridSpec(3,2)
plt.subplot(3,2,1)
plt.plot(x,y_sin)
plt.subplot(3,2,2)
plt.plot(x,y_cos)
plt.subplot(grid[2:3,:])
plt.plot(y_sin,y_cos)
```

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=20)
plt.title('Histogram')
plt.show()
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data, vert= True , patch_artist = True);
```
